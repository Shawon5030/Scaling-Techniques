```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


```

# loading csv file

```python
df=pd.read_csv('price.csv')
x=df[['Price']]
x
```

# minmax_scaler


```python
scaler=MinMaxScaler()
```

```python
df['MinMaxScaler']=scaler.fit_transform(df[['Price']])
```

```python
df
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler_1=StandardScaler()
df['StanderScaler']=scaler_1.fit_transform(x)
df
```

# Robust scaler

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler_3=RobustScaler()
df['RobustScaler']=scaler_3.fit_transform(x)
```

# Log Transform

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np

sceler_4=FunctionTransformer(np.log1p)
df['Log']=sceler_4.fit_transform(x)
df
```

# box cox

```python
from scipy.stats import boxcox
```

```python
x_=(df['Price']+1)
df['box_cox'], _ =boxcox(x_)
```

```python
df
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
scaler_6=MaxAbsScaler()
x=df[['Price']]
df['MaxAbsScaler']=scaler_6.fit_transform(x)
df
```

```python
df.to_csv('Scaling Techniques.csv')
```
